–ó–∞–¥–∞–Ω–∏–µ 1
–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª—É—á–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π numpy.array X –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∞—Å—Å–∏–≤ –∏–∑ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –º–∞—Å—Å–∏–≤–∞. 

```python
def cumulative_mean(X):
    # —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–≤–Ω–æ: cumsum X –¥–µ–ª–∏—Ç—å –Ω–∞ –º–∞—Å—Å–∏–≤ –æ—Ç 1 –¥–æ –¥–ª–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ X
    result = np.array(X.cumsum()/np.arange(1,len(X)+1))
    return result
```

```python
cumulative_mean(a)
```

```python
# test
a = np.array([1, 2, 4, -9])

np.allclose(cumulative_mean(a), np.array([1, 1.5, 7./3, -0.5]))
```

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –±–∏–±–ª–∏–æ—Ç–µ–∫—É numpy (–≤–∞–º –ø–æ–º–æ–∂–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è cumsum).

–í—ã—Ö–æ–¥–æ–º —Ñ—É–Ω–∫—Ü–∏–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –¥–≤—É–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤ —Ç–æ–π –∂–µ —Ñ–æ—Ä–º—ã, —á—Ç–æ –∏ X

–∫–æ—Ç–æ—Ä–∞—è —Ä–∞–±–æ—Ç–∞–µ—Ç –Ω–µ —Ç–æ–ª—å–∫–æ —Å –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–º–∏, –Ω–æ –∏ —Å –¥–≤—É–º–µ—Ä–Ω—ã–º–∏ –º–∞—Å—Å–∏–≤–∞–º–∏.

```python
import numpy as np

def cumulative_mean_2d(A):
    """  
    A: np.array[num_row, num_column]        --- –º–∞—Ç—Ä–∏—Ü–∞-–∞—Ä–≥—É–º–µ–Ω—Ç
    return S: np.array[num_row, num_column] --- –≤—ã—Ö–æ–¥–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö —Å—É–º–º

    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –º–∞—Ç—Ä–∏—Ü—É A —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é n x m –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç 
    –º–∞—Ç—Ä–∏—Ü—É —Å —Ç–æ–π –∂–µ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é n x m, i-–∞—è —Å—Ç—Ä–æ—á–∫–∞ –∫–æ—Ç–æ—Ä–æ–π –µ—Å—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å 
    –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö —Å—É–º–º —ç–ª–µ–º–µ–Ω—Ç–æ–≤ i-–æ–π —Å—Ç—Ä–æ–∫–∏ –º–∞—Ç—Ä–∏—Ü—ã A

    –í —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ —ç—Ç–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª –ø–∞–∫–µ—Ç–∞ numpy

    """ 
    #YOUR CODE

    result = np.array(A.cumsum(axis=1)/np.arange(1,len(A[0])+1))
    return result
```

```python
#test
a = np.array([1, 2, 4, -9, 1, 1, 1, 1]).reshape(2, 4)

np.allclose(cumulative_mean_2d(a), 
                   np.array([1, 1.5, 7./3, -0.5, 1, 1, 1, 1]).reshape(2,4))
```

```python
def transform(X):
    X[1::2]=1
    X[::2]=X[::2]*X[::2]*X[::2]
    result = X
    return result
```

```python
#test
X = np.array([5, -2, 44, 8, 0, -2, -3, 1, 5, 3])

np.allclose(transform(X), np.array([125, 1, 85184, 1, 0, 1, -27, 1, 125, 1]))
```

```python
import numpy as np #–Ω–µ —Å—Ç–∏—Ä–∞—Ç—å!        

def diag_2k(a):
    b=np.diagonal(a)
    c = b[b%2==0]
    #param a: np.array[size, size]
    #YOUR CODE

    result = sum(c)
    return result
```

```python
#test
a = np.array([1, 3, 6, 
              2, 8,-2, 
              -2, 0, -2], 
             dtype=int).reshape(3,3)

np.allclose(diag_2k(a), 6)
```

```python
def no_numpy_scalar(v1, v2):
    
    """  
    v1: list[n] --- –ø–µ—Ä–≤—ã–π –≤–µ–∫—Ç–æ—Ä-–∞—Ä–≥—É–º–µ–Ω—Ç –¥–ª–∏–Ω–æ–π n
    v2: list[n] --- –≤—Ç–æ—Ä–æ–π –≤–µ–∫—Ç–æ—Ä-–∞—Ä–≥—É–º–µ–Ω—Ç –¥–ª–∏–Ω–æ–π n
    return c: float  --- —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å–∫–∞–ª—è—Ä–Ω–æ–≥–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–æ–≤ v1 –∏ v2

    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–∞ –≤–µ–∫—Ç–æ—Ä–∞ –¥–ª–∏–Ω–æ–π n
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ, —Ä–∞–≤–Ω–æ–µ –∏—Ö —Å–∫–∞–ª—è—Ä–Ω–æ–º—É –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—é v1 –∏ v2

    –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤, –Ω–µ –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –ø–∞–∫–µ—Ç–∞ numpy

    """ 
    result = v1@v2
    return result
```

```python
def numpy_scalar(v1, v2):
    """  
    v1: np.array[n] --- –ø–µ—Ä–≤—ã–π –≤–µ–∫—Ç–æ—Ä-–∞—Ä–≥—É–º–µ–Ω—Ç –¥–ª–∏–Ω–æ–π n
    v2: np.array[n] --- –≤—Ç–æ—Ä–æ–π –≤–µ–∫—Ç–æ—Ä-–∞—Ä–≥—É–º–µ–Ω—Ç –¥–ª–∏–Ω–æ–π n
    return c: float  --- —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å–∫–∞–ª—è—Ä–Ω–æ–≥–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è –≤–µ–∫—Ç–æ—Ä–æ–≤ v1 –∏ v2

    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–∞ –≤–µ–∫—Ç–æ—Ä–∞ –¥–ª–∏–Ω–æ–π n
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ, —Ä–∞–≤–Ω–æ–µ –∏—Ö —Å–∫–∞–ª—è—Ä–Ω–æ–º—É –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—é

    –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –ø–∞–∫–µ—Ç–∞ numpy
    """ 
    result = np.dot(v1,v2)
    return result
```

```python
def no_numpy_mult(first, second):
    first=np.array(first)
    second=np.array(second)
    """
    A: list of "size" lists, each contains "size" floats --- –ø–µ—Ä–≤–∞—è –º–∞—Ç—Ä–∏—Ü–∞-–∞—Ä–≥—É–º–µ–Ω—Ç
    B: list of "size" lists, each contains "size" floats --- –≤—Ç–æ—Ä–∞—è –º–∞—Ç—Ä–∏—Ü–∞-–∞—Ä–≥—É–º–µ–Ω—Ç
    return C: list of "size" lists, each contains "size" floats --- –º–∞—Ç—Ä–∏—Ü–∞, —è–≤–ª—è—é—â–∞—è—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º    —É–º–Ω–æ–∂–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü a –∏ b

    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–µ –º–∞—Ç—Ä–∏—Ü—ã: A –∏ B —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é size x size
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É –∏—Ö –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è A * B = C 

    –†–µ–∞–ª–∏–∑—É–π—Ç–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–π –∏–∑ –ø–∞–∫–µ—Ç–∞ numpy
    """
    #YOUR CODE: please do not use numpy

    result = list(first@second)
    return result


def numpy_mult(first, second):
    """
    A: np.array[size, size]              --- –ø–µ—Ä–≤–∞—è –º–∞—Ç—Ä–∏—Ü–∞-–∞—Ä–≥—É–º–µ–Ω—Ç
    B: np.array[size, size]              --- –≤—Ç–æ—Ä–∞—è –º–∞—Ç—Ä–∏—Ü–∞-–∞—Ä–≥—É–º–µ–Ω—Ç
    return C: np.array[size, size]       --- –º–∞—Ç—Ä–∏—Ü–∞, —è–≤–ª—è—é—â–∞—è—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º —É–º–Ω–æ–∂–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü A –∏ B

    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–µ –º–∞—Ç—Ä–∏—Ü—ã: A –∏ B —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å—é size x size
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É –∏—Ö –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è A * B = C 

    –†–µ–∞–ª–∏–∑—É–π—Ç–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ –ø–∞–∫–µ—Ç–∞ numpy    """

    #YOUR CODE: please use numpy

    result = np.dot(first,second)
    return result
    
```

```python
n = X.shape[1]
r = np.linalg.matrix_rank(X)
```

```python
r
```

```python
U, sigma, VT = np.linalg.svd(X, full_matrices=False)
```

```python
D_plus = np.diag(np.hstack([1/sigma[:r], np.zeros(n-r)]))
```

```python
V = VT.T
```

```python
X_plus = V.dot(D_plus).dot(U.T)
```

```python
w = X_plus.dot(y)
```

```python
error = np.linalg.norm(X.dot(w) - y, ord=2) ** 2
error
```

```python
def calculate_loss_function(X, w, y):
    n = X.shape[1]
    r = np.linalg.matrix_rank(X)

    U, sigma, VT = np.linalg.svd(X, full_matrices=False)
    D_plus = np.diag(np.hstack([1/sigma[:r], np.zeros(n-r)]))
    
    V = VT.T
    X_plus = V.dot(D_plus).dot(U.T)
    
    error = np.linalg.norm(X.dot(w) - y, ord=2)**2
    
    return error
```

```python
# –í—Å–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –≤ —Å–∂–∞—Ç–æ–º –≤–∏–¥–µ
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# –î–∞–Ω–Ω—ã–µ
X = data.drop(columns='Churn')
y = data['Churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25)

# –ú–æ–¥–µ–ª—å
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
preds_valid = knn.predict(X_valid)

# –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
print(f'Accuracy={accuracy_score(y_valid, preds_valid)}')
print(f'Precision={precision_score(y_valid, preds_valid)}')
print(f'Recall={recall_score(y_valid, preds_valid)}')
```

–£ –Ω–∞—Å –Ω–µ—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤, —á—Ç–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –æ–±–ª–µ–≥—á–∞–µ—Ç –∑–∞–¥–∞—á—É.

```python
import pandas as pd
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# –î–∞–Ω–Ω—ã–µ

X = data.drop(columns='Churn')
y = data['Churn']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25)
```

–ü–æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å –Ω–æ–≤—ã–º —Ç–∏–ø–æ–º –æ–±—ä–µ–∫—Ç–æ–≤ –≤ sklearn: Transformer. –£ –Ω–∏—Ö –µ—Å—Ç—å –¥–≤–∞ –æ—Å–Ω–æ–≤–Ω—ã—Ö –º–µ—Ç–æ–¥–∞ .fit() –∏ .transform(). –≠—Ç–∏ –æ–±—ä–µ–∫—Ç—ã –∫–∞–∫–∏–º-—Ç–æ –æ–±—Ä–∞–∑–æ–º —É—á–∞—Ç—Å—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –∏ –ø–æ—Ç–æ–º –º–æ–≥—É—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ. –í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ, Standard Scaler –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ –∑–∞–ø–æ–º–∏–Ω–∞–µ—Ç —Å—Ä–µ–¥–Ω–µ–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞. –ü–æ—Å–ª–µ –æ–Ω –Ω–æ—Ä–º–∏—Ä—É–µ—Ç –¥–∞–Ω–Ω—ã–µ, –∏—Å–ø–æ–ª—å–∑—É—è –∏—Ö. 

$$x_{new} = \frac{x-mean}{std}$$

–í–æ–æ–±—â–µ, –Ω–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å—Ç–æ–∏—Ç –ø–æ—á—Ç–∏ –≤—Å–µ–≥–¥–∞. –≠—Ç–æ –æ—á–µ–Ω—å –≤–∞–∂–Ω–æ –¥–ª—è
* –ú–µ—Ç—Ä–∏—á–µ—Å–∫–∏—Ö –∞–ª–æ–≥—Ä–∏—Ç–º–æ–≤
* –õ–∏–Ω–µ–π–Ω—ã—Ö –∞–ª–æ–≥—Ä–∏—Ç–º–æ–≤ (–º—ã –µ—â–µ —É–∑–Ω–∞–µ–º, —á—Ç–æ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ö–æ—Ä–æ—à–æ –Ω–µ —Ç–æ–ª—å–∫–æ –æ—Ç–Ω–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å, –Ω–æ –∏ –∫–∞–∫ –º–æ–∂–Ω–æ —Å–∏–ª—å–Ω–µ–µ –ø—Ä–∏–±–ª–∏–∑–∏—Ç—å –∏—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É). –°—é–¥–∞ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫–∞–∫ –≤—Å–µ–≤–æ–∑–º–æ–∂–Ω—ã–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, —Ç–∞–∫ –∏ PCA, –∫–æ—Ç–æ—Ä—ã–π –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –Ω–∏–∂–µ.

–ù–æ—Ä–º–∏—Ä–æ–≤–∫–∞ –Ω–µ –≤–∞–∂–Ω–∞ —Ä–∞–∑–≤–µ —á—Ç–æ –¥–ª—è –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã—Ö –Ω–∞ —Ä–µ—à–∞—é—â–∏—Ö –¥–µ—Ä–µ–≤—å—è—Ö (—Å–ª–µ–¥—É—é—â–µ–µ –∑–∞–Ω—è—Ç–∏–µ).

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train, y_train)
X_train_normed = scaler.transform(X_train)
X_valid_normed = scaler.transform(X_valid)
```

```python
# –ü–æ–∂–Ω–µ–º –ø–ª–æ–¥—ã –∏ –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å
```

```python
# –ú–æ–¥–µ–ª—å
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_normed, y_train)
preds_valid = knn.predict(X_valid_normed)

# –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ
print(f'Accuracy={accuracy_score(y_valid, preds_valid)}')
```
