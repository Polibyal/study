```python
data = pd.DataFrame(
{
"X_1": ["1", "1", "1", "0", "0", "0", "0", "1"],
"X_2": ["0", "0", "0", "1", "0", "0", "0", "1"],
"Y": ["1", "1", "1", "1", "0", "0", "0", "0"],
}
)
```

```python
import math

lst_p1, lst_p2, lst_E = [], [], []
for i in np.arange(0.01, 1.0, 0.01):
  p1 = i
  p2 = 1-i
  lst_p1.append(p1)
  lst_p2.append(p2)
  E = -(p1 * math.log(p1,2) + p2 * math.log(p2,2))
  lst_E.append(E)

E_data = pd.DataFrame({'p1':lst_p1, 'p2':lst_p2, 'E':lst_E})

sns.lineplot(data=E_data, x='p1', y='E')
plt.title('Энтропия для 2-х классов')
```

```python
def ent(cats=0, dogs=0):
    p_cats = cats/(cats+dogs)
    p_dogs = dogs/(cats+dogs)
    sx1 = -p_cats * math.log2(p_cats) if p_cats else 0
    sx2 = -p_dogs * math.log2(p_dogs) if p_dogs else 0
    return sx1 + sx2

print('Шерстист=0 branch entropy', ent(cats=1))
print('Шерстист=1 branch entropy', ent(cats=5, dogs=4))
print('Гавкает=0 branch entropy', ent(cats=5))
print('Гавкает=1 branch entropy', ent(cats=1, dogs=4))
print('Лазает=0 branch entropy', ent(dogs=4))
print('Лазает=1 branch entropy', ent(cats=6))
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Загрузка данных и разделение на признаки и целевую переменную
iris = load_iris()
X = iris.data
y = iris.target

# Создание объекта решающего дерева
dt = DecisionTreeClassifier()

# Определение множества параметров для перебора
param_grid = {
    'max_depth': range(1, 11),
    'min_samples_split': range(2, 11),
    'min_samples_leaf': range(1, 11)
}

# Создание объекта GridSearchCV для перебора параметров
search = GridSearchCV(dt, param_grid, cv=5)

# Обучение GridSearchCV на данных ириса
search.fit(X, y)

# Сохранение лучшего дерева в переменную best_tree
best_tree = search.best_estimator_
```
